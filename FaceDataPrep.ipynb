```python
import os
from PIL import Image
```

```python
folder_path = "./RealWorldOccludedFaces/images/neutral/"
```

```python
directory_names = [
        name
        for name in os.listdir(folder_path)
        if os.path.isdir(os.path.join(folder_path, name))
    ]
```

```python
color_path='./Data/Color_Images/'
i=1
for name in directory_names:
  
    print(f'Copying image of: {name}')
    # Open the image using PIL
    source_path=folder_path+name+'/000001.jpg'
    print(source_path)
    img = Image.open(source_path)
    destination_path=color_path+'Image'+str(i)+'.png'
    # Save the image to the destination
    print(destination_path)
    img.save(destination_path,"JPEG")
    i=i+1
    print(f'Image{i}')

    

    
    
```
