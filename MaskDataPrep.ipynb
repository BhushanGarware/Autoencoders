```python
import os
from PIL import Image
```

```python
folder_path = "./RealWorldOccludedFaces/images/masked/"
```

```python
directory_names = [
        name
        for name in os.listdir(folder_path)
        if os.path.isdir(os.path.join(folder_path, name))
    ]
```

```python
i=1
for mask_name in directory_names:
    try:
        print(f'Copying image of: {mask_name}')
        # Open the image using PIL
        mask_source_path=folder_path+mask_name+'/000001.jpg'
        print(mask_source_path)
        mask_img = Image.open(mask_source_path)
        
        face=mask_name.split('_wearing')[0]
        print(face)
        face_source_path='./RealWorldOccludedFaces/images/neutral/'+face+'/000001.jpg'
        print(face_source_path)
        face_img = Image.open(face_source_path)

        destination_mask='./Data/noisy_images/'+str(i)+'.png'
        destination_face='./Data/clean_images/'+str(i)+'.png'
        # Save the image to the destination 
        mask_img.save(destination_mask)
        face_img.save(destination_face)
        print(f'Image{i}')
        i=i+1
    except:
        print('No Mask')
    
        
    

    
    
```
